```python
from os import path

data_dir = "../ud120-projects/final_project/"
ed_pickle_path = path.join(data_dir, "final_project_dataset.pkl")
ed_names_path = path.join(data_dir, "poi_names.txt")
```

```python
import pickle

enron_data = pickle.load(open(ed_pickle_path, "rb"))
```

```python
import pandas as pd
```

```python
def to_comma_name(name):
    parts = name.split()
    if len(parts) > 1 and len(parts[1]) < 3:
        parts.pop(1)
    last = parts[0].capitalize()
    first = parts[1].capitalize() if len(parts)>1 else ""
    return f'{last}, {first}'

data = pd.DataFrame.from_dict(enron_data, orient='index')
data['name'] = data.index.map(to_comma_name)
data
```

```python
len(data[data.poi == True])
```

```python
with open(ed_names_path, 'rt') as names_file:
    lines = names_file.readlines()
    pois = set([line[4:].strip() 
           for line in lines
           if line.startswith("(")
    ])
pois, len(pois)
```

```python
poi_data = data[data.name.isin(pois)]
poi_data
```
